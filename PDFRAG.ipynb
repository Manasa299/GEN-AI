```python
#import libraries
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader(
    file_path = "E://GEN AI//The Ultimate Python Handbook.pdf"
)
```

```python
pages=loader.load()#by default it splits based upon the pages
```

```python
print(pages)
```

```python
print(len(pages))
```

```python
!pip install -U langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_documents(pages)
```

```python
print(len(chunks))
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embedder= GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview")
```

```python
!pip install -qU langchain-qdrant
```

```python
from langchain_qdrant import QdrantVectorStore
from qdrant_client import QdrantClient
```

```python
vector_store = QdrantVectorStore.from_documents(
    documents=[],
    url="http://localhost:6333",
    collection_name="demo_collection",
    embedding=embedder,
)
```

```python
vector_store.add_documents(documents=chunks)
print("Inhection part done")
```

```python
retrive=QdrantVectorStore.from_existing_collection(
    embedding=embedder,
    collection_name="pdf_collection",
    url="http://localhost:6333")
```

```python
query="What is Python?"
found_docs=retrive.similarity_search(query)
found_docs
```
